```python
brush = alt.selection_interval() 
# All = [gold] + [iron] + [nickel] + [silver]
input_dropdown = alt.binding_select(options=['All' ,'gold','iron','nickel','silver'], name='Assets')
selection = alt.selection_single(fields=['Assets'], bind=input_dropdown)
chart = alt.Chart(df).mark_point().encode( 
    x='yearmonth(Dates):T',
    y = 'deposit',  color = alt.condition(brush, 'Tr_type', alt.value('lightgray')),
    tooltip = 'Balance').add_selection(
  selection
).add_selection(
    brush
).transform_filter(
    selection
).interactive()
chart.encode(y='deposit') | chart.encode(y='Balance')
```
